## Pandas

A partir do dataset `movies.csv` utilize o Pandas para responder as seguintes perguntas:

### Qual o gênero de filme mais lucrativo em todo o mundo?

```python
# ADICIONE O SEU CÓDIGO AQUI

import pandas as pd
#Maneira global de remover a notação científica
pd.options.display.float_format = '{:20,.2f} $'.format
movies_df = pd.read_csv("movies.csv")
wwg = movies_df.groupby("Major Genre").mean()
wwg["Genre Profit"] = wwg["Worldwide Gross"] - wwg["Production Budget"]
wwg = wwg.sort_values(by="Genre Profit", ascending = False)[["Worldwide Gross", "Production Budget", "Genre Profit"]];wwg
```

### Qual o diretor dirigiu mais filmes?

```python
#Top 5
directed_films = movies_df.groupby("Director").count()
directed_films = directed_films.sort_values(by="Title", ascending = False)[["Title"]]
directed_films = directed_films.rename(columns = {'Title': "Directed_Films"});directed_films.iloc[0:5]
```

### Quais filmes tiverem os maiores arrecadações (Gross) em relação ao orçamento de produção (Budget)?

```python
#top 10
lucrative = movies_df.groupby("Title").mean()
lucrative["Lucrative"] = lucrative["Worldwide Gross"] - lucrative["Production Budget"]
lucrative = lucrative.sort_values(by="Lucrative", ascending = False)[["Lucrative"]];lucrative.iloc[0:10]
```
