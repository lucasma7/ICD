## Correlação

Colete os seguintes dados dos times do campeonato brasileiro de 2020:

- Pontos
- Gols marcados
- Gol sofridos

Você pode encontrar esses dados no site da CBF (www.cbf.com.br)

Calcule a correlação entre:

- Pontos e gols marcados
- Pontos e gols sofridos

```python
import pandas as pd
import numpy as np

brasileiro_2020 = pd.read_html("https://www.cbf.com.br/futebol-brasileiro/competicoes/campeonato-brasileiro-serie-a/2020")
brasileiro_2020 = np.array([brasileiro_2020][0], dtype=object)
brasileiro_2020 = pd.DataFrame(brasileiro_2020[0]);brasileiro_2020
```

### Correlação entre Pontos e gols marcados

```python
brasileiro_2020[["PTS","GP"]].corr()
```

### Correlação entre Pontos e gols sofridos

```python
brasileiro_2020[["PTS","GC"]].corr()
```

### Análise dos Resultados

Qual o nível (fraco, moderado, forte) das correlações entre as variáveis analisadas?

**Com relação aos pontos e gols marcados, temos uma correlação muito forte, sendo quase 1.
Em relação aos pontos e gol sofridos temos uma anticorrelação forte.**

De acordo com os dados analisados, é melhor um time focar em marcar mais gols ou sofrer menos gols?

**Segundo os dados, é melhor o time focar em realizar mais gols, já que a correlação é muito forte, chegando muito proximo de 1**
