## Elo Ranking

```python
import math
import pandas as pd
import numpy as np

def logit(x, c=0.4):
    return 1/(1+math.exp(-c*x)).astype(float)
```

```python
def prob(a, b):
    return logit(a-b)
```

```python
def mi(a, b):
    return 1*prob(a,b) + ((-1)*(1-prob(a,b)))
```

```python
def update_ranking(a, b, result, k):
    return a + k*(result-mi(a,b))
```

```python
movies_df = pd.read_csv("movies.csv")
movies_df
```

Para termos uma classificação mais convicta, iremos unir as colunas do Rotten Tomatoes Rating(a ecala é de 0-100) e IMDB Rating, soma ela e duvidir por dois

```python

movies_rating = movies_df
movies_rating = movies_rating.groupby("Title").mean()
movies_rating["Median_Rating"] = (((movies_rating["Rotten Tomatoes Rating"]/10) +  movies_rating["IMDB Rating"])/2) - (movies_rating["Production Budget"]/10e7)
movies_rating["Lucrative"] = movies_rating["Worldwide Gross"] - movies_rating["Production Budget"]
movies_r = movies_rating.sort_values(by="Median_Rating", ascending = False)[[ "Rotten Tomatoes Rating", "IMDB Rating",
                                                                             "Median_Rating", "Lucrative","Distributor"]]



```
